# Creating new columns amp_or_non_amp , story and position from AD_UNIT_NAME

```python
dfp['AD_UNIT_NAME'].unique()
```

```python
def get_story(x):
    if 'amp' in x:
        story = None
        x_list = x.split('-')
        pos = x_list[2].split(' ')[0]
        amp =  'Amp'
    if 'story' in x:
        x_list = x.split('_')
        story = x_list[2]
        pos = x_list[5].split(' ')[0]
        amp = 'Nonamp'
    
    return story, pos, amp
        
```

```python
t = dfp['AD_UNIT_NAME'].apply(get_story).tolist()
```

```python
t
```

```python
df_with_new_cols=pd.DataFrame(t,columns=['story','position','amp_or_non_amp'])
```

```python
df_with_new_cols.head()
```

```python
df_with_new_cols.shape
```

```python
dfp.shape
```

```python
df_merged=pd.merge(dfp, df_with_new_cols, left_index=True, right_index=True)
```

```python
df_merged.head()
```

```python
df_merged.isna().sum()
```

```python
df_merged['DAY'].unique()
```

# mapping the Day column to MONDAY, TUESDAY, WEDNESDAY,THURSDAY,FRIDAY, SATURDAY and SUNDAY

```python
df_merged['DAY']=df_merged['DAY'].replace(to_replace=[1, 2, 3, 4, 5, 6, 7], value=['MONDAY','TUESDAY','WEDNESDAY','THURSDAY','FRIDAY','SATURDAY','SUNDAY'])
        
```

# Filtering out blanks from Actual_eCPM. Filtered out data will be used for merging the datasets

```python
actualECPM['Actual_eCPM']=actualECPM['Actual_eCPM'].replace(to_replace='-',value='None')
```

```python
actualECPM['Actual_eCPM'].unique()
```

```python
df_filtered_with_actualCPM=actualECPM[actualECPM['Actual_eCPM']!='None']
```

```python
df_filtered_with_actualCPM
```

```python
df_merged1=pd.merge(df_merged,df_filtered_with_actualCPM,on='LINE_ITEM_NAME', how='left')
```

```python
df_merged1
```

# Actual_eCPM is created. 
To treat NaNs we will use np.where and will obtain final Actual_eCPM

```python
df_merged1.Actual_eCPM = np.where(df_merged1.Actual_eCPM.isnull(), df_merged1.eCPM, df_merged1.Actual_eCPM)
```

```python
df_merged1.dtypes
```

```python
df_merged1['Actual_eCPM']=df_merged1['Actual_eCPM'].astype(float)
```

```python
df_merged1
```

```python
df_merged1['Actual_Revenue']=df_merged1['Impressions']*df_merged1['Actual_eCPM']
```

```python
df_merged1.info()
```

```python
df_merged1['Actual_Revenue']=df_merged1['Actual_Revenue']/1000
```
